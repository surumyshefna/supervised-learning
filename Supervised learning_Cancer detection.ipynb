### Summary:
- **Precision**: Focuses on the accuracy of positive predictions.
- **Recall**: Focuses on identifying all true positive cases.
- **F1 Score**: Balances precision and recall in a single metric.

These metrics are useful in understanding the performance of your classification model, particularly in cases where the class distribution is skewed or when false positives and false negatives carry different costs.

```python
# Function to store evaluation metrics in a dictionary
def get_performance_metrics(model_name,tr_accuracy,test_accuracy,precision,recall,f1,confusion):
    return {
        'Model': model_name,
        'Train Accuracy': tr_accuracy,
        'Test Accuracy': test_accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Confusion Matrix':confusion
    }
performance_data=[]
```

**Confusion Matrix (Logistic Regression)**

```python
L_confusion=confusion_matrix(x_test_predict,y_test)
print("Confusion Matrix is")
print(L_confusion)
```

```python
performance_data.append(get_performance_metrics("Logistic Regression",L_train_accuracy,L_test_accuracy,L_precision,L_recall,L_f1,L_confusion))
```
